```python
def get_leaves(G):
    leaves = []
    for node in G.nodes:
        if(G.degree(node) == 1):
            leaves += [node]
    return leaves
```

```python
G = nx.Graph()
G.add_edges_from([('a','b'),('a','d'),('c','d')])
```

```python
assert set(get_leaves(G)) == {'c','b'}
```

```python
print(get_leaves(G))
```
